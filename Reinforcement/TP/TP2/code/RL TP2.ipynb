## Help Functions

```python
def print_means(means):
    
    tb_means = PrettyTable()
    N = len(means)

    tb_means.add_column("index", range(1, N + 1))
    tb_means.add_column("mean", np.around(means, 3))
    
    print(tb_means)
```

```python
def plot_regrets(curves, labels, title = "Expected Reward w.r.t. Simulation Time"):
    
    plt.figure()
    for i in range(len(curves)):
        plt.plot(curves[i], label = labels[i])
    plt.legend()
    plt.xlabel("t")
    plt.ylabel("Regret")
    plt.title(title)
    plt.grid()
    plt.show()
```

```python
def plot_errors(curves, labels, title = "L2 norm of |θ* - θt| w.r.t. Simulation Time"):
    
    plt.figure()
    for i in range(len(curves)):
        plt.plot(curves[i], label = labels[i])

    plt.legend()
    plt.xlabel("t")
    plt.ylabel("L2 norm")
    plt.title(title)
    plt.grid()
    plt.show()
```

## Stochastic Multi-Armed Bandits on Simulated Data

### Bernoulli Bandit Models

In the module **algos_smab** three simulation strategies are implemented, namely:

* UCB1 algorithm

$$
A_{t+1} = argmax_{a \in \{1, \cdots, K\}} \hat{\mu}_a (t) + \rho_t \sqrt{\cfrac{\log(t)}{2 N_a (t)}}
$$

* Thompson Sampling

$$
A_{t+1} = argmax_{a \in \{1, \cdots, K\}} \theta_a (t) \quad \mbox{ where } \theta_a(t) \sim \pi_a(t)
$$


* Naive strategy

$$
A_{t+1} = argmax_{a \in \{1, \cdots, K\}} \hat{\mu}_a (t)
$$

The function *expected_regret* calculates the expected regret of each strategy with respect to the simulation time $t$.

The function *complexity* calculates $C(p) = \sum_{a:p_a < p^{\star}} \cfrac{p^\star - p_a}{kl(p_a, p^\star)}$, the lower bound of $\cfrac{R_T}{log(T)}$.

#### Q1 

In the first example, we generate a series of bandits whose complexity is rather high.

```python
K = 10
K_means = np.random.rand(K) / 2
# Create bandits instances
MAB = [arms.ArmBernoulli(K_means[i], np.random.randint(1, 312414)) for i in range(K)]
compx = algos_smab.complexity(MAB)

print("The number of bandits: %d" % K)
print("\nThe means of bandits:")
print_means(K_means)
print("\nThe complexity of MAB: %f" % compx)
```

```python
N, T = 100, 10000

print("Simulation for UCB1...")
ucb = algos_smab.expected_regret(N, T, MAB, "ucb1")
print("Simulation for TS...")
ts = algos_smab.expected_regret(N, T, MAB, "thompson")
print("Simulation for Naive...")
naive = algos_smab.expected_regret(N, T, MAB, "naive")

oracle = compx * np.log(np.arange(1, T + 1))
```

```python
plot_regrets([ucb, ts, naive, oracle], ["UCB1", "Thompson", "NAIVE", "ORACLE"])
```

### Non-parametric Bandits
#### Q2

The UCB1 algorithm can be used directly on non-parametric MABs, however the Thompson Sampling cannot, thus the algorithm needs to be modified.

Each time when we sample from a chosen bandit the reward $r_i$, we simulate from a Bernoulli distribution of mean $r_i$ a new binary reward $\hat{r}_i$, which will be used as the reward of timestamp $i$.

From the [article](https://arxiv.org/pdf/1407.4443.pdf), we deduce that the notion of complexity still make sense.

```python
K = 10
K_means = np.random.rand(K, 2)
# Create bandits instances
MAB = [arms.ArmBeta(K_means[i, 0], K_means[i, 1], np.random.randint(1, 312414)) for i in range(K)]
compx = algos_smab.complexity(MAB)

print("The number of bandits: %d" % K)
print("\nThe means of bandits:")
print_means(K_means)
print("\nThe complexity of MAB: %f" % compx)
```

```python
N, T = 100, 10000

print("Simulation for UCB1...")
ucb = algos_smab.expected_regret(N, T, MAB, "ucb1")
print("Simulation for TS...")
ts = algos_smab.expected_regret(N, T, MAB, "thompson", True)
print("Simulation for Naive...")
naive = algos_smab.expected_regret(N, T, MAB, "naive")

oracle = compx * np.log(np.arange(1, T + 1))
```

```python
plot_regrets(ucb, ts, naive, oracle)
```

## Linear Bandit on Real Data

```python
from linearmab_models import ToyLinearModel, ColdStartMovieLensModel
from tqdm import tqdm
import algos_linmab
```

```python
random_state = np.random.randint(0, 24532523)
model = ColdStartMovieLensModel(random_state = random_state, noise = 0.1)
```

```python
model = ToyLinearModel(
    n_features = 8,
    n_actions = 20,
    random_state = random_state,
    noise = 0.1)
```

### LinUCB

#### The influence of $\alpha$

First we observe the influence of $\alpha$. This parameter controls the upper confidence bound. 

```python
lalpha = [0.1, 0.5, 1, 2, 3, 5, 10]
N, T = 50, 6000

lerrors, lregrets = [], []

for alpha in lalpha:
    print("Simulation for alpha = %.2f..." % alpha)
    lin_error, lin_regret = algos_linmab.optim_oracle(model, N, T, "LinUCB", alpha = alpha)
    lerrors.append(lin_error)
    lregrets.append(lin_regret)
    
plot_errors(lerrors, ["Alpha %f" % alpha for alpha in lalpha])
plot_regrets(lregrets, ["Alpha %f" % alpha for alpha in lalpha])
```

We conclude from the results that the l2 error between $\theta$ and $\theta^\star$ decreases when $\alpha$ is larger, however, the cumulative regret increases. Thus we need to keep a balance between this two criterions. In the following experiment, we use $\alpha = 2$.

#### The influence of $\lambda$

$\lambda$ is the regularization term when we estimate $\theta$.

```python
llambd = [0.1, 0.5, 1, 3, 5]
lerrors, lregrets = [], []

for lambd in llambd:
    print("Simulation for lambda = %.2f..." % lambd)
    lin_error, lin_regret = algos_linmab.optim_oracle(model, N, T, "LinUCB", alpha = 2, lambd = lambd)
    lerrors.append(lin_error)
    lregrets.append(lin_regret)
    
plot_errors(lerrors, ["Lambda %f" % lambd for lambd in llambd])
plot_regrets(lregrets, ["Lambda %f" % lambd for lambd in llambd])
```

We observe from the results that the l2 error between $\theta$ and $\theta^\star$ increases when $\lambda$ becomes smaller, however, the cumulative regret decreases. Thus we choose $\lambda = 1$ to keep a balance between these two criterions.

### Greedy Model

#### The influence of $\epsilon$

$\epsilon$ controls the possibility we pick a random action from the set. 

```python
leps = [0.05, 0.1, 0.2, 0.3, 0.5, 0.8]
lerrors, lregrets = [], []

for eps in leps:
    print("Simulation for eps = %.2f..." % eps)
    greedy_error, greedy_regret = algos_linmab.optim_oracle(model, N, T, "Greedy", epsilon = eps)
    lerrors.append(greedy_error)
    lregrets.append(greedy_regret)
    
plot_errors(lerrors, ["Epsilon %f" % eps for eps in leps])
plot_regrets(lregrets, ["Epsilon %f" % eps for eps in leps])
```

We observe from the results that the l2 error between $\theta$ and $\theta^\star$ increases when $\epsilon$ becomes smaller, however, the cumulative regret decreases. Thus we choose $\epsilon = 0.2$ to keep a balance between these two criterions.

### Model Comparison

```python
print("Simulation for LinUCB...")
lin_error, lin_regret = algos_linmab.optim_oracle(model, N, T, "LinUCB", alpha = 2, lambd = 1)
print("Simulation for Random Policy...")
random_error, random_regret = algos_linmab.optim_oracle(model, N, T, "Random", lambd = 1)
print("Simulation for Greedy Policy...")
greedy_error, greedy_regret = algos_linmab.optim_oracle(model, N, T, "Greedy", lambd = 1, epsilon = 0.2)

plot_errors([lin_error, random_error, greedy_error], ["LinUCB", "RANDOM", "GREEDY"])
plot_regrets([lin_regret, random_regret, greedy_regret], ["LinUCB", "Random", "Greedy"])
```
